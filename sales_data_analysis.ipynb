```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
              'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Sales': [25000, 27000, 30000, 35000, 37000, 40000,
              38000, 42000, 41000, 45000, 48000, 50000],
    'Profit': [8000, 8500, 9500, 11000, 11500, 12000,
               11800, 12500, 12300, 13500, 14000, 14500],
    'Marketing Spend': [4000, 4200, 4300, 4700, 4800, 5000,
                        4900, 5100, 5050, 5300, 5400, 5500]
}
df = pd.DataFrame(data)
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(df['Month'], df['Sales'], marker='o', color='teal')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales in INR')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(10,5))
sns.barplot(x='Month', y='Profit', data=df, hue='Month', palette='pastel', legend=False)
plt.title('Monthly Profit')
plt.xlabel('Month')
plt.ylabel('Profit in INR')
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Sales', y='Profit', data=df, hue='Month', palette='viridis', s=100)
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['Sales', 'Profit', 'Marketing Spend']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```
